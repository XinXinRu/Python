```python
#python可以免輸入等於去輸出資料
a = 7
b = 1
a+b
```

```python
#「import」套用「calendar」物件
import calendar

#僅要單月份
#輸出2021年2月的日曆
print(calendar.month(2021,2))
```

```python
'''當所需的物件有被「import」則免需再次import'''

#要整份日曆
#印出2022年的日曆
print(calendar.calendar(2022))
```

```python
#產生http請求，下載網頁資料
import requests

#a去盛裝「requests」要求「get」得到"括弧"內的資料
a = requests.get("https://www.youtube.com")

#印出a的資料及格式
print(a.text)
```

```python
#「webbrowser」開啟&瀏覽網頁
import webbrowser

b = requests.get("https://random.dog/woof.json").json()["url"]
webbrowser.open(b)
```

```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [4,5,6]

#將x及y的資料放至「plt.bar」
'''資料圖表為「bar」長直圖'''
plt.bar(x,y)

#印出來
plt.show()
```

```python
#印出圓餅圖

#資料大小為x，標示為y
plt.pie(x, labels=y)
plt.show()
```
