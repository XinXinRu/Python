eval函式的輸入是"字串str"型態的數學式，回傳則是數學式計算的結果

```python
#input()跳出一小行讓使用者可以輸入
#「input()」括弧內可以輸入要請使用者輸入的內容(如為文字)則需要加「" "」
#!但input出來的內容為文字，若要數值則需使用eval()!
#如float(eval(input()))這樣的型態轉換較為安全
#「eval()」可以將括弧內的內容轉換成數值
car_dist= float(eval(input("請輸入以公里為單位的行駛距離："))) #float(eval(input("請輸入以公里為單位的行駛距離")))，先將input進的內容(String型態)轉(eval)為數值再轉為浮點數(float)
run_hour= int(eval(input("請輸入經過的小時數："))) #int(eval(input("請輸入經過的小時數")))，先將input進的內容(String型態)轉(eval)為數值再轉為整數(int)
run_min= int(eval(input("請輸入經過的分鐘："))) #int(eval(input("請輸入經過地分鐘")))，先將input進的內容(String型態)轉(eval)為數值再轉為整數(int)
run_time= run_hour + run_min/60 #計算，「/」為除法
car_speed= car_dist / run_time #計算，「/」為除法
print("汽車的時速是：", car_speed, "km/hr") #顯示
```

```python
def reverse_str(oStr): #讀取oStr的內容
  reStr="" #每次讀取到的字放進去 #原先設定為空值
  for index in range(len(oStr)-1, -1, -1): #(1) #讀取oStr的內容，用range可以單獨字讀取「range(開始，結束，前進)」
    reStr += oStr[index] #(2) #每次讀取到的字放進去
  return reStr #回傳給reStr
print(reverse_str("Madam I'm Adam")) #顯示
```

(for迴圈與if判斷式)

```python
def count_word(word, words_list): #def用來判斷 count_word(word, words_list)，word是輸入想要用來判斷words_list內的英文單字是否有word所要的英文字，words_list判斷一串的英文單字)
  count=0 #出現的次數
  for keyword in words_list: #keyword是用來抓word_list內的單字
    if keyword == word: #判斷keyword跟word是否一樣
      count+=1 #如果是的話count+1
  return count #回傳count的數量
words_list=['i','am','a','boy','you','are','a','girl']
word=input("你想要查詢哪個單字的數量?")
wtotal=count_word(word, words_list)
print("在文章中一共有':", wtotal,"個單字", word)
```

```python
def do_times(n1,n2): #判斷 do_times(n1,n2) 使用者可以輸入兩個值(文字or數字)
  if (n1 is None or n2 is None): #如果輸入的第一個是「None」或輸入的第二個是「None」，結果為「真」的話
    return("你少填了被乘數與乘數") #上行為真時，回傳「你少填了被乘數與乘數」
  if (n1==0 or n2==0): #如果輸入的第一個是「0」或輸入的第二個是「0」，結果為真的話
    return("你的乘數或被乘數為0") #上行為真時，回傳「你的乘數或被乘數為0」
  else: #若上述的判斷都沒過的話則執行此
    return n1*n2 #回傳第一個及第二個輸入的值相乘
print(do_times(None,2))
print(do_times(2,0))
print(do_times(2,3))
```

```python
#在撰寫程式時會經常發生錯誤，為了避免發生錯誤時一行一行找建議使用「try:程式寫的地方except:錯誤顯示」
try: 
  file1=open("file1.txt", 'r') #開檔案來讀取
  file2=open("file2.txt", 'w+') #開檔案來寫入，若存在則覆蓋 
except IOError: #當出現錯誤時
  print("檔案不存在") #印出「檔案不存在」
else: #否則
  i=1 #變數i為1
  for line in file1: #讓line裡面的值傳進file1內做查詢
    print(line.rstrip()) #印出line裡面指定删除字串尾部指定的字
    file2.write("line " + str(i) + ":" + line) #file2寫入內容，通常在檔尾
    i=i+1 #可以縮寫為，i+=1
  file1.close() #檔案關閉
  file2.close() #檔案關閉
```
